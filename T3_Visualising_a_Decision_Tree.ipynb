```python
# Importing libaries required for data handling
import pandas as pd
```

```python
# Getting the DATA
df=pd.read_csv("https://raw.githubusercontent.com/Kuntal-G/Machine-Learning/master/R-machine-learning/data/banknote-authentication.csv")
```

Here i am Using " Bank Note Authentication UCI data " 

Data were extracted from images that were taken from genuine and forged banknote-like specimens. For digitization, an industrial camera usually used for print inspection was used. The final images have 400x 400 pixels. Due to the object lens and distance to the investigated object gray-scale pictures with a resolution of about 660 dpi were gained. Wavelet Transform tool were used to extract features from images.

FOR MORE IMFORMATION ON DATA

https://archive.ics.uci.edu/ml/datasets/banknote+authentication#

```python
df.head()
```

Class "0" represent the :- FAKE BANK NOTE Class

Class "1" represent the :- REAL BANK NOTE Class

```python
# Creating a decision tree classifier for the giving data
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,4].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=0)
```

```python
dtree=DecisionTreeClassifier()
dtree.fit(x_train,y_train)

print('Decision Tree Classifer Created')
```

```python
# CHECKING THE ACCURACY OF THE MODEL
y_pred=dtree.predict(x_test)
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
```

Mean Absolute Error of < 0.01 shows high level of accuracy
